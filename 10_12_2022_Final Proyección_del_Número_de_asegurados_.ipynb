## Transformación de tipo de datos

## Importando pycaret

```python
from pycaret.time_series import TSForecastingExperiment
```

```python
data.Fecha = data.datetime
```

```python
data.info()

```

```python
data.drop(columns=["Año","Mes","dia","Trimestre","Dia_semana","Semana","Dia_año","datetime"],inplace=True)
```

```python
data.index = data.Fecha
data.drop(columns="Fecha",inplace=True)
```

```python
data.info()
```

```python
y = data
```

```python
fh = 24 
fold = 3
```

## Configurando parametros para modelo de series de tiempo

```python
eda = TSForecastingExperiment()
eda.setup(data=y, fh=fh, fold=fold, session_id=42)
```

```python
eda.plot_model()  
```

```python
eda.plot_model(data_kwargs={"plot_data_type": ["original", "imputed", "transformed"]})
```

```python
eda.plot_model(plot="acf")
```

```python
eda.plot_model(plot="pacf", data_kwargs={'nlags':36}, fig_kwargs={'height': 500, "width": 800})
```

```python
eda.plot_model(plot="periodogram")
eda.plot_model(plot="fft")
```

```python
eda.plot_model(plot="diagnostics", fig_kwargs={"height": 800, "width": 1000})
```

```python
eda.plot_model(plot="decomp")
```

```python
eda.plot_model(plot="train_test_split", fig_kwargs={"height": 400, "width": 900})
```

##**Modelos**##

```python
eda.models()
```

##**Generacion de Modelo Naive**##

```python
model = eda.create_model("naive")
y_predict = eda.predict_model(model)
y_predict
eda.plot_model(estimator=model)
eda.plot_model(model, plot='insample')
model = eda.create_model("lightgbm_cds_dt")
y_predict = eda.predict_model(model)
eda.plot_model(estimator=model)
```

##**Optimizando**##

```python
tuned_model = eda.tune_model(model)
eda.plot_model(estimator=tuned_model)
print(model)
print(tuned_model)
eda.plot_model([model, tuned_model], data_kwargs={"labels": ["Baseline", "Tuned"]})
```
